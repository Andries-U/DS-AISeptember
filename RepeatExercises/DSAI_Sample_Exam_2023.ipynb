## Question 4

The diameter of pistons produced in a factory are _D_ cm where _D_ ~ Nor(13.4, 0.12)

(1) The tolerance specified for the pistons is that the diameter needs to be at least 13.35 cm and not more than 13.5 cm. What proportion of the production in the factory meets these tolerance limits?  
(2) Three pistons are chosen at random. What is the probability that none of them meet the tolerance limits  
(3) A sample of 20 pistons is taken at random. The mean value for the diameter is found to be 13.43 cm. Test at the 5% significance level whether any modifications have been made to the machine producing the pistons.


```python
sigma = 0.12
mean = 13.4
significance = 0.05


n = 20

percentage= (stats.norm.cdf(13.5, loc=mean, scale=sigma) - stats.norm.cdf(13.35, loc=mean, scale=sigma))

print("1. ", percentage)
print("2. ", (1-percentage)**3)



g1 = stats.t.isf(1-(significance/2), loc= mean, scale=sigma,df=n-1)
g2 = stats.t.isf((significance/2),loc= mean, scale=sigma, df=n-1)

print("3. the test mean is between within the 95% interval and we do not reject H0. [", g1, ",",g2, "]")

```

```python
import scipy.stats as stats

mean = 13.4
std_dev = 0.12

# Calculate Z-scores for 13.5 and 13.35
z_13_5 = (13.5 - mean) / std_dev
z_13_35 = (13.35 - mean) / std_dev

# Calculate cumulative probabilities
p_13_5 = stats.norm.cdf(z_13_5)
p_13_35 = stats.norm.cdf(z_13_35)

# Calculate the proportion
proportion_within_limits = p_13_5 - p_13_35

print("Proportion within tolerance limits:", proportion_within_limits)
```

```python
p_not_within_limits = (1 - (p_13_5-p_13_35))

p_none_within_limits = p_not_within_limits ** 3

print("Probability that none meet tolerance limits:", p_none_within_limits)
```

```python
import numpy as np

sample_mean = 13.43
sample_size = 20
sample_std_dev = 0.12  # Assuming the sample standard deviation is the same as the population

# Calculate the t-statistic
t_statistic = (sample_mean - mean) / (sample_std_dev / np.sqrt(sample_size))

# Calculate the critical t-value for a two-tailed test at 5% significance level
alpha = 0.05
degrees_of_freedom = sample_size - 1
t_critical = stats.t.ppf(1 - alpha/2, df=degrees_of_freedom)

# Perform the hypothesis test
if np.abs(t_statistic) > t_critical:
    print("Reject the null hypothesis: Modifications have been made.")
else:
    print("Fail to reject the null hypothesis: No evidence of modifications.")
```

To compare two groups of continuous data, a paired t-test (also known as a dependent t-test) is appropriate. This test will allow us to determine whether there is a statistically significant difference between the measurements at time1 and time2.

```python
# Perform a paired t-test
t_statistic, p_value = stats.ttest_rel(temperatures['time1'], temperatures['time2'])

print("Calculated t-statistic:", t_statistic)
print("Calculated p-value:", p_value)
```

To analyze whether customers have a distinct preference for any of the products, we can perform a chi-squared test of independence. This test will help us determine if there's an association between the products and the customers' choices.

Step 3: Formulate the hypotheses
Null Hypothesis (H0): The choice of product is independent of customer preferences (no distinct preference).
Alternative Hypothesis (H1): The choice of product is not independent of customer preferences (distinct preference).

```python
# Create a contingency table from the data
contingency_table = pd.crosstab(index=products['Choice'], columns='count', margins=True)
print(contingency_table)


# Perform the chi-squared test
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print("Calculated chi-squared statistic: ", chi2)

print("Calculated P-value: ", p_value)
```
