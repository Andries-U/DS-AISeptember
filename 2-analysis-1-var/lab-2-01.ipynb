# Lab 2.01 - Australian Athletes (ais)

Import all necessary Python libraries and load the AIS dataset from lab exercise 1.01. Also include the code for setting the index and conversion to categorical variables!

```python
# Importing the necessary packages
import numpy as np                                  # "Scientific computing"
import scipy.stats as stats                         # Statistical tests

import pandas as pd                                 # Data Frame
from pandas.api.types import CategoricalDtype

import matplotlib.pyplot as plt                     # Basic visualisation
from statsmodels.graphics.mosaicplot import mosaic  # Mosaic diagram
import seaborn as sns                               # Advanced data visualisation

# importing data
ais = pd.read_csv('https://raw.githubusercontent.com/HoGentTIN/dsai-labs/main/data/ais.csv')

# set index
ais.set_index(['id'])

# setting categories
ais.sex = ais.sex.astype(CategoricalDtype(categories=["f", "m"], ordered=False))

ais.sport = ais.sport.astype(CategoricalDtype(categories=['B_Ball', 'Row', 'Netball', 'Swim', 'Field', 'T_400m', 'T_Sprnt','Tennis', 'Gym', 'W_Polo'], ordered=False))

sns.countplot(data=ais, x='sport')
```

Select the following subsets from the dataset and calculate for each the appropriate measures of central tendency (and, when possible, dispersion) of the variables `ht` and `sex`:

1. the rowers
2. the rowers, netball and tennis players together
3. the female basketball players and rowers together

```python
ais_rower = ais[ais.sport=="Row"]

ais_row_net_tennis = ais[ais.sport.isin(["Row", "Tennis","Netball"])]

ais_female_basket_row = ais[(ais.sex == "f") & ais.sport.isin(["Row","B_Ball"])]

print("Rowers")
print("\nht".upper())
print(ais_rower.ht.describe())
print("\nsex".upper())
print(ais_rower.sex.describe())

print("\n" + "*"*50 + "\n")

print("Rowers, Netball and Basketball")
print("\nht".upper())
print(ais_row_net_tennis.ht.describe())
print("\nsex".upper())
print(ais_row_net_tennis.sex.describe())

print("\n" + "*"*50 + "\n")

print("Female Basketball and rowers")
print("\nht".upper())
print(ais_female_basket_row.ht.describe())
print("\nsex".upper())
print(ais_female_basket_row.sex.describe())



#print("mean ht = ", ais_rower.ht.mean())
```
